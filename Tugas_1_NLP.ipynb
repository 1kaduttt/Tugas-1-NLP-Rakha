```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import joblib

# 1. DATA ACQUISITION (Simulasi Data)
# Kamu bisa ganti ini dengan load file CSV kalau punya
data = {
    'text': [
        'aplikasinya keren banget membantu', 'sangat puas belanja di sini',
        'bagus sekali layanannya', 'kurirnya ramah', 'barang sampai dengan aman',
        'kecewa banget barang rusak', 'pengiriman sangat lambat',
        'aplikasi sering error', 'pelayanan buruk', 'tidak rekomendasikan'
    ],
    'label': [
        'Positif', 'Positif', 'Positif', 'Positif', 'Positif',
        'Negatif', 'Negatif', 'Negatif', 'Negatif', 'Negatif'
    ]
}
df = pd.DataFrame(data)

# 2. DATA PREPROCESSING & FEATURE ENGINEERING
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df['text'])
y = df['label']

# 3. MODELLING (Training)
model = MultinomialNB()
model.fit(X, y)

# 4. EVALUATION (Simulasi sederhana)
print("Model berhasil dilatih!")

# 5. EXPORT MODEL (Ini yang paling penting untuk Streamlit)
joblib.dump(model, 'sentiment_model.pkl')
joblib.dump(tfidf, 'tfidf_vectorizer.pkl')

print("File .pkl berhasil diunduh ke folder Colab!")
```
